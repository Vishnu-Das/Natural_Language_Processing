# TF-IDF
## Term Frequency - Inverse Document Frequency

We lower all the words in sentence and then we remove all stop words

### Term Frequency

#### TF = (Frequency of word in a sentence) / (Total words in a sentence}

### Inverse Document Frequency

#### IDF = loge (Number of sentences / Number of sentences containing the word)

#### Final TF-IDF = TF * IDF



Now we have got a vector matrix (sentences against vocabs)

### Advantage
- Intuitive
- Fixed size vocab
- **Word importance is getting captured**

### Disadvantages
- Sparse matrix still exist
- Out of vocabulary
