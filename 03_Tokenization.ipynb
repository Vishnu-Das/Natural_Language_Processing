```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
corpus = """Hello, how are you doing today? I hope you're having a great day!. Please let me know if you need any help with your project or Vishnu's project."""
```

```python
from nltk.tokenize import sent_tokenize
```

```python
print(sent_tokenize(corpus))
```

```python
documents = sent_tokenize(corpus)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Tokenization using word_tokenize
from nltk.tokenize import word_tokenize
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
words = []
for sentence in documents:
    words.extend(word_tokenize(sentence))
print(words)
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
for sentence in documents:
    print(wordpunct_tokenize(sentence))
```

```python
from nltk.tokenize import TreebankWordTokenizer # Does not treat sull stop as separate tokens. Akthough last full stop is treated as separate token.
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
