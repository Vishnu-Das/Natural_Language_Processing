# Bag Of Words

## Steps
- Lower case all the words
- Stop words
- calculate the vocabulary and the frequency for each word
- frequency should be ordered ijn descending order. max heap
- each word in vocab is a feature, we can choose the top most features and even ignore the ones with frequency of 1.
- Now for each sentence we create a vector, mapping the top features to 1 and other words to 0 in sentence
- But when a word is repeated then the count in the vector matrix will become 2 for that word
- Binary BOW will make the count to 1 no matter what the value is, but in normal BOW the count will get updated based on howmany times the word comes in sentence

### Advantage
- Simple and intuitive
- Here the vector or inputs are getting fixed (Fixed size inputs)

### Dis Advantage
- Sparse matrix or array issue is still there and can lead to overfitting
- Very lesas semantic information capturewd because ordering of the words in a sentence gets changes since we are ordering them in matrix based on the frequency of the word in vocab
- Out of Vocabulary is also an issue (when a new word comes in the test data)
- Ordering of words is getting changed
- Cosin similarity issue
  - The food is good : [1 1 1 0 1]
  - The food is not good : [1 1 1 1 1]
  - Above sentences are totally opposite of each other but when vector comparison happens they might have cosin similarity that makes them close since opnly one dimension have the change in vector.
  - Which is a wrong assumption to make.
