# Lemmatization

Lemmatization is a technique similar to stemming. The output we get after is called **lemma**, which is a root word and not a root stem

## Wordnet Lemmatizaation

Thin wrapper around wordnet corpus

```python
from nltk import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("fairly"), lemmatizer.lemmatize("sportingly"), lemmatizer.lemmatize("going")
```

POS

- Noun - n
- Verb - v
- Adjective - a
- Adverb - r

```python
lemmatizer.lemmatize("fairly", pos="r"), lemmatizer.lemmatize("sportingly", pos="r"), lemmatizer.lemmatize("going", pos="v")
```

```python
words = ["eating", "eats", "eaten", "writing", "writes", "programming", "programs", "history", "finally", "finalize"]
```

```python
for word in words:
    print(f"{word} --> {lemmatizer.lemmatize(word)}")
```

```python
for word in words:
    print(f"{word} --> {lemmatizer.lemmatize(word, pos='v')}")
```

```python
for word in words:
    print(f"{word} --> {lemmatizer.lemmatize(word, pos='a')}")
```

```python
for word in words:
    print(f"{word} --> {lemmatizer.lemmatize(word, pos='r')}")
```

Lemmatization is going to take more time than stemming. since it uses a morphy() function to WordNet CorpusReader class to find a lemma.
