# Stemming
Stemming is process of reducing a word to its word stem that **affixes to suffixes and prefixes or ot the roots of the word** known as lemma.

For ex : [Going, goes, gone] here **go** is the stem, in [eating, eats, eaten] **eat** is the stem

# 
## Porter Stemmer

```python
words = ["eating", "eats", "eaten", "writing", "writes", "programming", "programs", "history", "finally", "finalize"]
```

```python
## PorterStemmer
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(f"{word} --> {word + " ----> " + stemming.stem(word)}")
```

When stemming is applied then some words can fully change the meaning
ex: history --> history ----> histori

```python
stemming.stem("congratulations")
```

## Regexp Stemmer

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer("ing$|s$|e$|able$", min=4)
```

```python
reg_stemmer.stem("eating")
```

```python
reg_stemmer.stem("ingeating")
```

## Snowball Stemmer
It performs better than portter stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snowball_stemmer = SnowballStemmer("english")
```

```python
for word in words:
    print(f"{word} --> {snowball_stemmer.stem(word)}")
```

```python
stemming.stem("fairly"), stemming.stem("sportingly")
```

```python
snowball_stemmer.stem("fairly"), snowball_stemmer.stem("sportingly")
```

major disadvantage of stemming is this only that it can produce wrong words. This problem is solved by **Lemmitization**
